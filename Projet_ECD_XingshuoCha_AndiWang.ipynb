```python
# 2.2 Préparation des données
# Pour analyse les données, il faut d'abord rincer les données.
# On analyse d'abord les variables qui sont intéresses.
# Les variable qui sont la type de string peut-être moins intéresses, par exemple ID, Source, Description.
# Et les variables qui signfient les localisation d'accidents ne sont pas intéresses car toutes les conditions du rue où l'accident se passe sont exprimées dans les variables Amenity,Bump,Crossing... 
# Donc on pense que les localisations precises servent à rien quand on sait les variables ci-dessus.
# Et la même raison pour ne sélectionne pas les variables Sunrise_Sunset,Civil_Twilight,Nautical_Twilight,Astronomical_Twilight quand on sait le temps.
```

```python
# Apès les sélections des variables, on a choisi les variables suivantes pour analyser:
variables = ['Start_Time','Distance(mi)','Temperature(F)','Wind_Chill(F)','Humidity(%)','Pressure(in)','Visibility(mi)','Wind_Speed(mph)','Precipitation(in)',
          'Amenity','Bump','Crossing','Give_Way','Junction','No_Exit','Railway','Roundabout','Station','Stop','Traffic_Calming','Traffic_Signal','Turning_Loop']
print(len(variables))
# Et la variable cible est Severity
cible = ['Severity']
```

```python
# On obtiens les données intéresses:
df = df_origin[variables]
target = df_origin[cible]
print(df.dtypes)
print(df.shape)
df.describe(include='all')
```

```python
# Pour la variable Start_Time,on ne garde que les heurs où l'accident se passe.
pd.set_option('mode.chained_assignment', None)
df.loc[:,'Start_Time'] = df['Start_Time'].map(lambda x:int((x[-8:-6])))
# Ici il y a 2 façons de transformer les valeur boolean: {True:1,False:0},{True:1,False:-1}
# Pour grandir les différences entre la valeur boolean True et False,on choisi la deuxième façon.
# Pour les variables boolean, on les transforme en 1/-1
boolean_variables = ['Amenity','Bump','Crossing','Give_Way','Junction','No_Exit','Railway','Roundabout','Station','Stop','Traffic_Calming','Traffic_Signal','Turning_Loop']
mapping = {True:1,False:-1}
for bv in boolean_variables:
    df[bv] = df[bv].map(lambda x:mapping[x])
df.head(10)
```

```python
# Pour les variables qui ont les valeurs manquantes, on choisi que les remplir avec les moyennes.
# Mais s'il y a une variable dont toutes les lignes sont manquantes ou sont les même, on l'enleve car cette variable a rien information.
import numpy as np
for variable in df.columns:
    if len(df[variable].value_counts()) == 1:
        df = df.drop(columns=variable)
        print('toutes les lignes de la variable {0} sont les même, on l\'enleve'.format(variable))
        continue
    value = df[variable].count()
    if value < len(df[variable]):
        if value == 0:
            df = df.drop(columns=[variable])
            print('toutes les lignes de la variable {0} sont manquantes, on l\'enleve'.format(variable))
        else:
            df[variable] = df[variable].fillna(df[variable].mean())
            print('remplir les lignes manquantes de la variable {0} avec la moyenne'.format(variable))

#Il n'y a plus des valeurs manquantes
df.head(10)
```

```python
# Alors il faut aussi normaliser les variables.
# On rencontre une question ici car il y a plusieur méthodes pour normalisation
# par exemple la méthode 'min-max' qui fait Z = Z-min/max-min
# par contre la méthode 'z-score' qui fait Z-espérance/type-écart
# On choisi la méthode 'min-max'
#df_normalise = (df - df.min()) / (df.max() - df.min())
#df_normalise.head(10)
# On choisi la méthode 'z-score'
from scipy.stats import zscore
df_normalise = df.copy()
df_normalise[columns_numerique] = df[columns_numerique].apply(zscore)
df_normalise.head(10)
```

```python
# Donc pour supprimer les anomalies, on choisi que supprime les données où la valeur des variables numériques ont supérieur 3 fois type-écart par rapport à l'espérance.
import numpy as np
from scipy.stats import zscore
# On ne garde que les points où les variables numériques ont inférieur 3 fois type-écart par rapport à l'espérance.
row_normal = (np.abs(zscore(df[columns_numerique])) < 3).all(axis=1)
df = df[row_normal]
target = df_origin[row_normal][cible]
# On refait les normalisations 
df_normalise = df.copy()
df_normalise[columns_numerique] = df[columns_numerique].apply(zscore)
df_normalise.head(10)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=8)
pca.fit(df_normalise)
print(pca.explained_variance_ratio_,sum(pca.explained_variance_ratio_))
# On fait l'algorithme PCA pour faire la réduction de dimension
# On peut voir que quand on sélecte les 8 premières composants,l'inertie est 85%.
# Donc on peut réduire les données 21 dimensions à 8 dimensions sans trop de perte. 
```

```python
# les coordonnées des individus après projeter sur les composants.
df_reduce = PCA(n_components=8).fit_transform(df_normalise)
df_reduce = pd.DataFrame(df_reduce)
df_reduce.columns = ['composant1','composant2','composant3','composant4','composant5','composant6','composant7','composant8']
df_reduce = df_reduce.set_index(df_normalise.index)
# les directions des variables après projeter sur les composants.
variable_reduce = np.transpose(pca.components_[:, :])
# Petite remarque: df.iloc[x,y] est de trouver une valeur selon les localisation xième linge,yième columne dans array
#            Mais df.loc[x,y] est de trouver une valeur selon index x de la df.
```

```python
import matplotlib.pyplot as plt
# les coordonnées des individus après projeter sur les 2 première composants.
premiere_composant = df_reduce.iloc[:,0].tolist()
deuxiere_composant = df_reduce.iloc[:,1].tolist()
# les directions des variables après projeter sur les 2 première composants.
variable_reduce_2premiere_composant = variable_reduce[:,0:2]
plt.figure(figsize=(6, 6), dpi=80)
plt.scatter(premiere_composant[:300],deuxiere_composant[:300],c = target[:300].to_numpy())

# afficher les directions des variables sur la première plan factorielle
for i in range(variable_reduce_2premiere_composant.shape[0]):
    plt.arrow(0, 0, variable_reduce_2premiere_composant[i,0]*4, variable_reduce_2premiere_composant[i,1]*4,color = 'b',alpha = 0.5)
    plt.text(variable_reduce_2premiere_composant[i,0]*4, variable_reduce_2premiere_composant[i,1]*4,df_normalise.columns[i], color = 'g', ha = 'center', va = 'center')
plt.axis([-2,5,-3,3])
plt.show()
```

```python
# On peut voir que les variables Humidity,Pressure a une relation avec la première composant, il correspond l'humidité de l'air.
# Wind_Chill,Temperature,Visibility et Start_Time a une relation inverse avec la première composant,il correspond la temperature de l'air et l'état de conducteur.
# Humidity,Wind_Chill,Distance a une relation avec la deuxière composant,il correspond l'humidité et refroidissement de l'air.
# Wind_Speed;Pressure,Visibility,Precipitation et Start_Time a une relation inverse avec la deuxière composant,il correspond la pressure de l'air.
```

```python
# 2.6 modèle prédictif "boîte blanche"  - Decision Tree
# L'objectif des modèles est de predire la valeur de Severity selon les autre variable.
# On peut construire une système en temps real,en entrant les variables L'heure actuel,Pressure,Wind Speed,Traffic Signal,etc de chaque route,
# sortant la valeur Severity de chaque route,Pour aider les gens prévenir les accidents graves.
# La variable cible est Severity,c'est plutôt une variable continus mais pas discrête,Donc les modèles sont plutôt pour une regression.
```

```python
# Pour réaliser une modèle prédictif,il faut d'abord déterminer les data train
# On observe la distribution de la variable cible, il y a moins de ligns où Severity = 1
print(target['Severity'].value_counts())
df_target1 = df_normalise[target['Severity']==1]
df_target2 = df_normalise[target['Severity']==2]
df_target3 = df_normalise[target['Severity']==3]
df_target4 = df_normalise[target['Severity']==4]

df_train = pd.DataFrame(columns = df_normalise.columns)
# On ajoute 800 individus qui a Severity = 1 dans le data train, 10000 individus pour l'autre valeur de Severity.
df_train = df_train.append(df_target1.iloc[:800,:])
df_train = df_train.append(df_target2.iloc[:10000,:])
df_train = df_train.append(df_target3.iloc[:10000,:])
df_train = df_train.append(df_target4.iloc[:10000,:])
target_train = [1 for i in range(800)]+[2 for i in range(10000)]+[3 for i in range(10000)]+[4 for i in range(10000)]
df_tar_train = pd.DataFrame(target_train,columns = cible)
print(df_tar_train.shape)
print(df_train.shape)
# On obtient 30933 individus pour faire validation d'un modèle.
```

```python
# En suite on créer le data test
df_test = pd.DataFrame(columns = df_normalise.columns)
df_test = df_test.append(df_target1.iloc[800:841,:])
df_test = df_test.append(df_target2.iloc[10000:11000,:])
df_test = df_test.append(df_target3.iloc[10000:11000,:])
df_test = df_test.append(df_target4.iloc[10000:11000,:])
target_test = [1 for i in range(41)]+[2 for i in range(1000)]+[3 for i in range(1000)]+[4 for i in range(1000)]
df_tar_test = pd.DataFrame(target_test,columns = cible)
print(df_tar_test.shape)
print(df_test.shape)
```

```python
# Il y a beaucoup de indicateur pour evaluer une modèle regression
# On choisi la neg_mean_squared_error qui sont la negatif de la MSE
import sklearn
print(sklearn.metrics.SCORERS.keys())
# Et éteindre la warning DataConversionWarning
import warnings
from sklearn.exceptions import DataConversionWarning
warnings.filterwarnings(action='ignore', category=DataConversionWarning)
```

```python
# On fait DecisionTree Regressor pour la modèle boître blanche
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_predict,cross_val_score
max_depth = [3,6,8,12,16]
for depth in max_depth:
    clf = DecisionTreeRegressor(max_depth = depth)
    print(np.mean(cross_val_score(clf, df_train, df_tar_train, cv=10,scoring = 'neg_mean_absolute_error')))
# Les scores des modèle sont neg_mean_squared_error
# On peut voir que en fur et à mesure que la max_depth d'un DecisionTree Ressgressor augmente,
# Le neg_mean_squared_error diminue d'abord et augmente ensuite, c'est à dire une sur apprentissage.
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_predict,cross_val_score
max_depth = 6
clf = DecisionTreeRegressor(max_depth = max_depth)
clf.fit(df_train, df_tar_train)
pre = clf.predict(df_test)
mae = 0
for i in range(len(pre)):
    mae = abs(df_tar_test.iloc[i]-pre[i])/len(pre) + mae
    print(int(df_tar_test.iloc[i]),pre[i])
print(mae)
# On peut vois que presque toutes les individus qui a Severity = 1 ne sont pas predire correctement,
# Car il y a juest 900 individus qui a Severity = 1 dans la data train
# Mais encore on peut voir la MAE de la modèle est 0.32, c'est à dire que chaque prediction de la modèle a une erreur ±0.32.
```

```python
# sklearn.svm.SVR
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_predict,cross_val_score
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
for kernel in kernels:
    svr = SVR(kernel = kernel)
    print(cross_val_score(svr, df_train, df_tar_train, cv=3,scoring = 'neg_mean_squared_error'))
# la performance de SVM est pas très bien.
```

```python
# On peut voir que la performace de RandomForest est similaire à Dicision Tree.
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_predict,cross_val_score
max_depth = [3,6,8,12,16]
for depth in max_depth:
    rf = RandomForestRegressor(max_depth = depth)
    print(np.mean(cross_val_score(rf, df_train, df_tar_train, cv=10,scoring = 'neg_mean_absolute_error')))
```

```python
# La modèle linéaire regressor a une performance qui a neg_mean_squared_error -0.68
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_predict,cross_val_score
lr = LinearRegression()
score = cross_val_score(lr, df_train, df_tar_train, cv=10,scoring = 'neg_mean_squared_error')
print(score)
print(np.mean(score))
```

```python
# La modèle Gradient Boosting regressor a une performance qui a neg_mean_squared_error -0.45,-0.47,-0.51
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import cross_val_predict,cross_val_score
n_estimators = [50,100,500]
for n_e in n_estimators:
    gbr = GradientBoostingRegressor(n_estimators = n_e)
    score = cross_val_score(gbr, df_train, df_tar_train, cv=10,scoring = 'neg_mean_squared_error')
    print(score)
    print(np.mean(score))
```

```python
# La modèle Gradient Boosting regressor a une performance qui a neg_mean_squared_error -0.72,-0.68,-0.65
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_predict,cross_val_score
n_neighbors = [5,10,20]
for n_n in n_neighbors:
    knn = KNeighborsRegressor(n_neighbors = n_n)
    score = cross_val_score(knn, df_train, df_tar_train, cv=10,scoring = 'neg_mean_squared_error')
    print(score)
    print(np.mean(score))
```

```python
# La modèle Neural Network regressor a une performance qui a neg_mean_squared_error -0.46,-0.65,-0.75
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import cross_val_predict,cross_val_score
hidden_layer_sizes = [(3,3,3),(3,3,3,3,3),(3,3,3,3,3,3,3,3,3,3)]
for size in hidden_layer_sizes:
    nw = MLPRegressor(hidden_layer_sizes = size,max_iter=500)
    score = cross_val_score(nw, df_train, df_tar_train, cv=10,scoring = 'neg_mean_squared_error')
    print(score)
    print(np.mean(score))
```
